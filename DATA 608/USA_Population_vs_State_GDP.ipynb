```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/nnaemeka-git/global-datasets/main/usa_population_gdp.csv'
ind = pd.read_csv(url)
```

```python
ind.head()
```

# Population vs GDP of U.S. States with Top 4 States Highlighted

```python
# Create a new variable 'per_capita' by dividing GDP by Population
ind['per_capita'] = (ind['GDP(in billion current U.S. dollars)'] * 1e9) / ind['Population']
```

```python
# Scatter plot of Population vs GDP with top 4 states circled
top_4_states = ind.nlargest(4, 'GDP(in billion current U.S. dollars)')
plt.figure(figsize=(12, 8))
plt.scatter(ind['Population'], ind['GDP(in billion current U.S. dollars)'], alpha=0.2)

# Highlight top 4 states
for i, row in ind.iterrows():
    if row['STATE'] in top_4_states['STATE'].values:
        plt.scatter(row['Population'], row['GDP(in billion current U.S. dollars)'], edgecolors='r', s=100, facecolors='none', linewidths=2)
    plt.text(row['Population'], row['GDP(in billion current U.S. dollars)'], row['CODE'], fontsize=9)

plt.title('Population vs GDP of U.S. States with Top 4 States Highlighted')
plt.xlabel('Population')
plt.ylabel('GDP (in billion current U.S. dollars)')
plt.grid(False)
plt.show()
```

# GDP Per Capita

```python
# Find the state with the highest GDP per capita
max_per_capita_state = ind.loc[ind['per_capita'].idxmax()]

# Scatter plot of per_capita vs CODE with the highest per_capita value circled
plt.figure(figsize=(10,5))
plt.scatter(ind['STATE'], ind['per_capita'], alpha=0.7, color='blue')

# Highlight the state with the highest GDP per capita
plt.scatter(max_per_capita_state['STATE'], max_per_capita_state['per_capita'], edgecolors='r', s=200, facecolors='none', linewidths=2)
plt.axhline(y = ind['per_capita'].mean(), xmin = 0, xmax = 10,color = "black", linestyle = "dashed")
plt.text(0, ind['per_capita'].mean(),"Average GDP Per Capita", color = "red")

plt.title('USA - GDP per Capita')
#plt.xlabel('State Code')
plt.ylabel('GDP per Capita (in current U.S. dollars)')
plt.grid(False)
plt.xticks(rotation=90)  # Rotate state codes for better readability

# Save the plot as an image file
plt.show()
```

To summarize the United States GDP: Approximately 17 states have a GDP per capita of around 65K the national average. A notable outlier is the District of Columbia, with a GDP per capita exceeding $200K. The top four states by GDPCalifornia (CA), Texas (TX), New York (NY), and Florida (FL)generally have a GDP per capita above the national average, except for Florida. Mississippi (MS) and West Virginia (WV) rank lower on the chart.
Data source: Population data was collected from the USA Census Bureau (https://www.census.gov/data/tables/time-series/demo/popest/2020s-state-total.html) and GDP figures were sourced from Statista (https://www.statista.com/statistics/248023/us-gross-domestic-product-gdp-by-state).
